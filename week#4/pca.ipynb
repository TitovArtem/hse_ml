```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
```

```python
close_prices = pd.read_csv('close_prices.csv')
```

```python
close_prices.head()
```

```python
train_y = close_prices.date
train_x = close_prices.drop('date', axis=1)
```

```python
train_x
```

```python
pca = PCA(n_components=10)
```

```python
pca.fit(train_x)
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.size
```

```python
pd.DataFrame(pca.components_)
```

```python
np.sum(pca.explained_variance_ratio_)
```

```python
for i in range(1, 10):
    temp_pca = PCA(n_components=i)
    temp_pca.fit(train_x)
    print i, np.sum(temp_pca.explained_variance_ratio_)
```

```python
pred_y = pca.transform(train_x)
```

```python
pd.DataFrame(pred_y).head()
```
