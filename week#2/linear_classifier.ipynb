```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
columns = ['y', 'x1', 'x2']
test = pd.read_csv('perceptron-test.csv', names=columns)
train = pd.read_csv('perceptron-train.csv', names=columns)
```

```python
test = test[test.y != -1]
train = train[train.y != -1]
```

```python
test.y = test.y.replace(0, -1)
train.y = train.y.replace(0, -1)
```

```python
test_x = test.drop('y', axis=1)
test_y = test.drop(['x1', 'x2'], axis=1)
train_x = train.drop('y', axis=1)
train_y = train.drop(['x1', 'x2'], axis=1)
```

```python
sc_test_x = StandardScaler().fit_transform(test_x)
sc_train_x = StandardScaler().fit_transform(train_x)
```

```python
def plot(x, y, w=None):
    if x.shape[0] != y.shape[0]:
        raise AttributeError('Invalid dimensions for x and y.')
    fig = plt.figure(figsize=(7, 7))
    colors = {1: 'r', -1: 'b'}
    for i in range(x.shape[0]):
        try:
            plt.plot(x[i][0], x[i][1], colors[y[i]] + 'o')
        except KeyError:
            continue
    plt.show()
    
```
